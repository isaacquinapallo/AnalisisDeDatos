```python
#Ejercicio 40
import numpy as np

# Tipos
for dtype in [np.int8, np.int32, np.int64]:
    print("Minimum representable value for {}: {}".format(dtype, np.iinfo(dtype).min))
    print("Maximum representable value for {}: {}".format(dtype, np.iinfo(dtype).max))

for dtype in [np.float32, np.float64]:
    print("Minimum representable value for {}: {}".format(dtype, np.finfo(dtype).min))
    print("Maximum representable value for {}: {}".format(dtype, np.finfo(dtype).max))
    print("Smallest positive representable number for {}: {}".format(dtype, np.finfo(dtype).eps))
```

```python
#Ejercicio 48
import numpy as np
X, Y = np.meshgrid(np.linspace(-1, 1, 10), np.linspace(-1, 1, 10))
D = np.sqrt(X**2 + Y**2)
sigma, mu = 1.0, 0.0

G = np.exp(-((D - mu)**2) / (2.0 * sigma**2))

print(G)
```
